```python
import sqlite3
import pandas as pd
```

```python
db_name = "bank.db"
```

```python
def select_from_table(query_select_table, db_name=db_name):
    connection = sqlite3.connect(db_name)
    cursor = connection.cursor()
    cursor.execute(query_select_table)
    df = pd.DataFrame(cursor.fetchall())
    return df
```

```python
select_from_table("""
    SELECT LoanTypeID, SUM(LoanAmount) AS TotalLoanAmount 
    FROM loans 
    GROUP BY LoanTypeID;
""")
```

```python
select_from_table("""
    SELECT BranchID, COUNT(DISTINCT CustomerID) AS CustomerCount 
    FROM customers 
    GROUP BY BranchID;
""")
```

```python
select_from_table("""
    SELECT CardType, UsageType, COUNT(*) AS CardCount 
    FROM cards 
    GROUP BY CardType, UsageType;
""")
```

```python
select_from_table("""
    SELECT TransactionChannel, SUM(Amount) AS TotalAmount 
    FROM transactions 
    GROUP BY TransactionChannel;
""")
```

```python
select_from_table("""
    SELECT l.CustomerID, 
        SUM(l.LoanAmount) AS TotalLoanAmount, 
        SUM(t.Amount) AS TotalTransactionAmount 
    FROM loans l 
    LEFT JOIN accounts a ON l.CustomerID = a.CustomerID 
    LEFT JOIN transactions t ON a.AccountID = t.AccountID 
    GROUP BY l.CustomerID;
""")
```

```python
select_from_table("""
    SELECT 
        (COUNT(CASE WHEN EWalletConnected = 'TRUE' THEN 1 END) * 100.0 / COUNT(*)) AS EWalletAdoptionRate 
    FROM accounts;
""")
```

```python
select_from_table("""
    SELECT BranchID, SUM(LoanAmount) AS TotalLoanAmount
    FROM loans
    GROUP BY BranchID
    ORDER BY TotalLoanAmount DESC;
""")
```

```python
select_from_table("""
    SELECT 
        (COUNT(DISTINCT t.AccountID) * 100.0 / COUNT(DISTINCT a.AccountID)) AS ActiveAccountsPercentage 
    FROM accounts a 
    LEFT JOIN transactions t ON a.AccountID = t.AccountID;
""")
```
