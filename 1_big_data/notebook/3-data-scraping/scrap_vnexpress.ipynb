```python
from bs4 import BeautifulSoup
import requests
import json
import re
```

```python
url = 'https://vnexpress.net'
```

```python
response_data = requests.get(url)
response_content = response_data.content
response_content
```

```python
soup = BeautifulSoup(response_content, 'html.parser') # json, xml
soup
```

```python
soup_head = soup.find_all('head')[0]
soup_head
```

```python
soup_body = soup.find_all('body')[0]
print(soup_body.prettify())
```

```python
soup_body_news = soup_body.find_all('h3', class_='title-news')
len(soup_body_news), soup_body_news
```

```python
href_list = []
for idx, news in enumerate(soup_body_news):
    print(idx)
#     print(news)
    print(news.find('a')['title'])
    print(news.find('a')['href'])
    print(news.find('a').text)
    print('\n')

    href_list.append(news.find('a')['href'])
```
