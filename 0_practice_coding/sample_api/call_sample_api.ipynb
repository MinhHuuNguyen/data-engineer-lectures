```python
import requests
```

```python
URL = 'https://reqres.in/api/users/'
```

```python
for user_id in range(10):
    user_url = URL + f'{user_id}'
    response = requests.get(user_url)

    print(user_url)
    print(response.json())
    print()
```

```python
post_data = {
    'name': 'Minhh',
    'job': 'Data Scientist'
}
```

```python
response = requests.post(user_url, data=post_data)

print(response.json())
```
