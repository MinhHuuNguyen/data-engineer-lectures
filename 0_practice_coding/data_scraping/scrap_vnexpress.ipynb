```python
href_list = []
for idx, news in enumerate(soup_body_news):
    print(idx)
#     print(news)
    print(news.find('a')['title'])
    print(news.find('a')['href'])
    print(news.find('a').text)
    print('\n')

    href_list.append(news.find('a')['href'])
```

```python
def scrap_data(url):
    response_data = requests.get(url)
    soup = BeautifulSoup(response_data.content, 'html.parser')
    soup_body = soup.find_all('body')[0]
    soup_page_detail = soup_body.find_all('div', class_='sidebar-1')

    soup_category = soup_page_detail[0].find_all('a', attrs={'data-medium': re.compile("Menu")})
    post_category = soup_category[0].text

    soup_date = soup_page_detail[0].find_all('span', class_="date")
    post_date = soup_date[0].text
    
    soup_title = soup_page_detail[0].find_all('h1', class_='title-detail')
    post_title = soup_title[0].text

    soup_description = soup_page_detail[0].find_all('p', class_="description")
    post_description = soup_description[0].text

    soup_content = soup_page_detail[0].find_all('p', class_="Normal")
    post_content = [_soup_content.text for _soup_content in soup_content]

    post_author =  post_content[-1]
    post_content.pop(-1)

    return {
        'url': url,
        'date': post_date,
        'title': post_title,
        'description': post_description,
        'content': post_content,
        'author': post_author
    }
```

```python
for url in href_list:
    results = scrap_data(url)

    pretty_json_str = json.dumps(results, indent=4, ensure_ascii=False)

    print(pretty_json_str)
    break
```
