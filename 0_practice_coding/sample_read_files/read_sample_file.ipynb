```python
# Python generator
with open('sample.txt', 'r') as file:
    line = file.readline()
    print(line)
    line = file.readline()
    print(line)
```

```python
file = open('sample.txt', 'r')

count = 0
while True:
    line = file.readline()
    print(line)
    count += 1

    if count == 3:
        break
```
