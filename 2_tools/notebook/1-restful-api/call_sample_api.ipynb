```python
import requests
```

```python
URL = 'https://reqres.in/api/users/'
```

```python
%%timeit
for user_id in range(5):
    user_url = URL + f'{user_id}'
    response = requests.get(user_url)
```

```python
for user_id in range(5):
    user_url = URL + f'{user_id}'
    response = requests.get(user_url)

    print(user_url)
    print(response.json())
    print()
```
